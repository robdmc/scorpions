```python
import easier as ezr
import pandas as pd
import lifelines as ll
%config Completer.use_jedi = False
```

```python
df = pd.read_csv('./scorpions_e1.csv')
df.columns = ezr.slugify(df.columns)
df.loc[:, 'trial'] = df.trial.str.strip()
df['dead'] = df.alive - 1


df.head()
```

```python
dfc = df[df.trial == 'C']
dfh = df[df.trial == 'H']
dfw = df[df.trial == 'W']

dfw.head()
```

```python
fit_control = ll.KaplanMeierFitter()
fit_control.fit(durations=dfc.treatment_days, event_observed=dfc.dead, label='Control')

fit_heat = ll.KaplanMeierFitter()
fit_heat.fit(durations=dfh.treatment_days, event_observed=dfh.dead, label='Heat')

fit_water = ll.KaplanMeierFitter()
fit_water.fit_left_censoring(durations=dfh.treatment_days, event_observed=dfh.dead, label='Water')

ax = ezr.figure()
ax = fit_control.plot_survival_function(ax=ax)
ax = fit_heat.plot_survival_function(ax=ax)
ax = fit_water.plot_survival_function(ax=ax)
ax.set_xlabel('Time')
ax.set_ylabel('Proportion Surviving')
```
